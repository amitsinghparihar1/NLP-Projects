```python
# get the independent feature
x=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
y.unique()
```

```python
y.replace(' is just unbelievably rich from North Sea oil. ','1', inplace=True)
```

```python
y.replace(' за участие. Благодарю вас.','1', inplace=True)
```

```python
y.value_counts()
```

```python
y=y.astype(int)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# Vocabulary_size
voc_size=5000
```

```python
message=x.copy()
```

```python
message.head()
```

```python
message.reset_index(inplace=True)
```

```python
nltk.download('stopwords')
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
# Dataset Preprocessing
from nltk.stem import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(message['title'])):
    review = re.sub('[^a-zA-Z]', ' ', message['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
 onehot_rpr=[one_hot(words,voc_size) for words in corpus]
```

```python
onehot_rpr
```

```python
# Embedding Representation
```

```python
sent_length = 20
embedded=pad_sequences(onehot_rpr, padding='pre',maxlen=sent_length)
```

```python
embedded
```

```python
# creating model
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
x_final = np.array(embedded)
y_final = np.array(y)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.3,shuffle=y,random_state=10)
```

```python
early_stopping= tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
# Model Training
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=30, batch_size=32, callbacks=early_stopping)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.where(y_pred>0.5,1,0)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('accuracy score of model is:',accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
